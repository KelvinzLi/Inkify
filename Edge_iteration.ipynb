```
import numpy as np
import cv2
import os

def cv_sobel(in_path, out_path):
  image = cv2.imread(in_path)
  image = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY) # to grey scale

  #Sobel Edge Detection
  sobelX = cv2.Sobel(image,cv2.CV_64F,1,0)
  sobelY = cv2.Sobel(image,cv2.CV_64F,0,1)

  sobelX = np.uint8(np.absolute(sobelX))
  sobelY = np.uint8(np.absolute(sobelY))

  sobelCombined = cv2.bitwise_or(sobelX,sobelY)#

  cv2.imwrite(out_path, sobelCombined)

def hed(in_path, out_path):
  os.system("python pytorch-hed/run.py --model bsds500 --in \'%s\' --out \'%s\'" %(in_path, out_path))
```

```
path=''
out_sobel=''
out_hed=''

path_list=os.listdir(path)

for name in path_list:
  print(name)

  path_name=os.path.join(path,name)
  path_sobel=os.path.join(out_sobel,name)
  path_hed=os.path.join(out_hed,name)

  cv_sobel(path_name, path_sobel)
  hed(path_name, path_hed)
```
